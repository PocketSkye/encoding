```python
def ordinal_encoding(data):
    unique_categories = sorted(set(data))
    
    ordinal_mapping = {}
    index = 1  
    for category in unique_categories:
        ordinal_mapping[category] = index
        index += 1
    
    encoded_data = []
    for category in data:
        encoded_data.append(ordinal_mapping[category]) 
    
    return encoded_data, ordinal_mapping

data = ['cat', 'dog', 'dog', 'cat', 'rabbit', 'cat', 'rabbit']

encoded_data, ordinal_mapping = ordinal_encoding(data)

print("Original Data: ", data)
print("Ordinal Mapping: ", ordinal_mapping)
print("Encoded Data: ", encoded_data)
```

```python
def one_hot_encoding(data):
    unique_categories = list(set(data))  
    one_hot_mapping = {}
    
    for category in unique_categories:
        one_hot_vector = [0] * len(unique_categories)  
        index = unique_categories.index(category) 
        one_hot_vector[index] = 1  
        one_hot_mapping[category] = one_hot_vector 
    
    encoded_data = []
    for category in data:
        encoded_data.append(one_hot_mapping[category]) 
    
    return encoded_data, unique_categories
encoded_data, categories = one_hot_encoding(data)

print("Original Data: ", data)
print("Categories: ", categories)
print("One-Hot Encoded Data: ")
for i in range(len(data)):
    print(f"{data[i]}: {encoded_data[i]}")
encoded_data
```
